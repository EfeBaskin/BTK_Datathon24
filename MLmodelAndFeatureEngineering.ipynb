````python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(dataframe):
    dataframe = dataframe.loc[:, dataframe.apply(pd.Series.nunique) != 1]

    X = sm.add_constant(dataframe)

    vif_data = pd.DataFrame()
    vif_data['Feature'] = dataframe.columns
    vif_data['VIF'] = [variance_inflation_factor(X.values, i + 1) for i in range(len(dataframe.columns))]

    return vif_data
vif_values = calculate_vif(test_data_reduced)
print(vif_values)
````

````python
# @title Capping Outliers (winsorizing)
from scipy import stats
import numpy as np

train_df_capped = train_data_reduced.copy()

numeric_columns = train_df_capped.select_dtypes(include=[np.number]).columns
threshold = 3
for i, column in enumerate(numeric_columns):
    mean_val = train_df_capped[column].mean()
    std_dev = train_df_capped[column].std()
    upper_bound = mean_val + threshold * std_dev
    lower_bound = mean_val - threshold * std_dev

    train_df_capped.loc[train_df_capped[column] > upper_bound, column] = upper_bound
    train_df_capped.loc[train_df_capped[column] < lower_bound, column] = lower_bound

original_shape = train_data_reduced.shape
new_shape = train_df_capped.shape
(original_shape, new_shape)
````

# **Creating new features**

````python
train_df_capped['Parental_Education_Sum'] = train_df_capped['Anne Egitim Durumu'] + train_df_capped['Baba Egitim Durumu']
test_data_reduced['Parental_Education_Sum'] = test_data_reduced['Anne Egitim Durumu'] + test_data_reduced['Baba Egitim Durumu']
````

````python
train_df_capped['Socioeconomic_Status'] = (train_df_capped['Anne Calisma Durumu']*1.5) + (train_df_capped['Baba Calisma Durumu']*1.5) * train_df_capped['Parental_Education_Sum']
test_data_reduced['Socioeconomic_Status'] = (test_data_reduced['Anne Calisma Durumu'] + test_data_reduced['Baba Calisma Durumu']) * test_data_reduced['Parental_Education_Sum']
````

````python
train_df_capped['Weighted_Involvement_Score'] = (train_df_capped['Stk Projesine Katildiniz Mi?'] * 2) + train_df_capped['Spor Dalindaki Rolunuz Nedir?'] + (train_df_capped['Girisimcilik Kulupleri Tarzi Bir Kulube Uye misiniz?'] * 2)
test_data_reduced['Weighted_Involvement_Score'] = (test_data_reduced['Stk Projesine Katildiniz Mi?'] * 2) + test_data_reduced['Spor Dalindaki Rolunuz Nedir?'] + (test_data_reduced['Girisimcilik Kulupleri Tarzi Bir Kulube Uye misiniz?'] * 2)
````

````python
# Trying to have the lowest RMSE possible.So I gave weights to certain columns which had biggest impact on our target column and made new ones using the existing ones
train_df_capped['Involvement_Girisimcilik'] = train_df_capped['Weighted_Involvement_Score'] * (train_df_capped['Girisimcilik Kulupleri Tarzi Bir Kulube Uye misiniz?'] *1.5)
test_data_reduced['Involvement_Girisimcilik'] = test_data_reduced['Weighted_Involvement_Score'] * (test_data_reduced['Girisimcilik Kulupleri Tarzi Bir Kulube Uye misiniz?'] *1.5)

train_df_capped['Social_Engagement_Score'] = train_df_capped['Aktif olarak bir STK üyesi misiniz?'] + \
                                               train_df_capped['Stk Projesine Katildiniz Mi?'] + \
                                               train_df_capped['Girisimcilik Kulupleri Tarzi Bir Kulube Uye misiniz?']
test_data_reduced['Social_Engagement_Score'] = test_data_reduced['Aktif olarak bir STK üyesi misiniz?'] + \
                                              test_data_reduced['Stk Projesine Katildiniz Mi?'] + \
                                              test_data_reduced['Girisimcilik Kulupleri Tarzi Bir Kulube Uye misiniz?']

train_df_capped['Parental_Influence_Score'] = train_df_capped['Parental_Education_Sum'] * train_df_capped['Socioeconomic_Status']
test_data_reduced['Parental_Influence_Score'] = test_data_reduced['Parental_Education_Sum'] * test_data_reduced['Socioeconomic_Status']

train_df_capped['Language_Education_Score'] = (train_df_capped['Ingilizce Biliyor musunuz?'] * 1.5) + train_df_capped['Parental_Education_Sum']
test_data_reduced['Language_Education_Score'] = (test_data_reduced['Ingilizce Biliyor musunuz?'] * 1.5) + test_data_reduced['Parental_Education_Sum']

train_df_capped['Active_Involvement'] = ((train_df_capped['Aktif olarak bir STK üyesi misiniz?'].astype(bool)) &
                                         (train_df_capped['Profesyonel Bir Spor Daliyla Mesgul musunuz?'].astype(bool))).astype(int)

test_data_reduced['Active_Involvement'] = ((test_data_reduced['Aktif olarak bir STK üyesi misiniz?'].astype(bool)) &
                                           (test_data_reduced['Profesyonel Bir Spor Daliyla Mesgul musunuz?'].astype(bool))).astype(int)



train_df_capped['Involvement_Parental_Interaction'] = train_df_capped['Weighted_Involvement_Score'] * train_df_capped['Parental_Education_Sum']
test_data_reduced['Involvement_Parental_Interaction'] = test_data_reduced['Weighted_Involvement_Score'] * test_data_reduced['Parental_Education_Sum']
````

````python
train_df_capped['Involvement_University_Interaction'] = train_df_capped['Active_Involvement'] * train_df_capped['Universite Not Ortalamasi']
test_data_reduced['Involvement_University_Interaction'] = test_data_reduced['Active_Involvement'] * test_data_reduced['Universite Not Ortalamasi']

university_freq = train_df_capped['Universite Adi'].value_counts(normalize=True)
train_df_capped['Universite_Adi_Freq'] = train_df_capped['Universite Adi'].map(university_freq)
test_data_reduced['Universite_Adi_Freq'] = test_data_reduced['Universite Adi'].map(university_freq)

train_df_capped['Application_Recency'] = 2024 - train_df_capped['Basvuru Yili']
test_data_reduced['Application_Recency'] = 2024 - test_data_reduced['Basvuru Yili']
train_df_capped.drop('Basvuru Yili', axis=1, inplace=True)
test_data_reduced.drop('Basvuru Yili', axis=1, inplace=True)

train_df_capped['Education_Bursary_Interaction'] = (train_df_capped['Anne Egitim Durumu'] + train_df_capped['Baba Egitim Durumu']) * train_df_capped['Burs Aliyor mu?']
test_data_reduced['Education_Bursary_Interaction'] = (test_data_reduced['Anne Egitim Durumu'] + test_data_reduced['Baba Egitim Durumu']) * test_data_reduced['Burs Aliyor mu?']

train_df_capped['Socioeconomic_Sibling_Impact'] = train_df_capped['Socioeconomic_Status'] * train_df_capped['Kardes Sayisi']
test_data_reduced['Socioeconomic_Sibling_Impact'] = test_data_reduced['Socioeconomic_Status'] * test_data_reduced['Kardes Sayisi']
````

````python
train_df_capped['Basvuru Yasi'] = train_df_capped['Basvuru Yasi'].fillna(0)
train_df_capped['Basvuru Yasi'] = train_df_capped['Basvuru Yasi'].astype(int)
````

# **Hyper Parameter Tuning with Optuna**
```
import optuna
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

def objective(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 100, 1500),
        'learning_rate': trial.suggest_float('learning_rate', 0.1, 0.5),
        'depth': trial.suggest_int('depth', 2, 10),
        'l2_leaf_reg': trial.suggest_int('l2_leaf_reg', 1, 10),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf', 10, 100),
        'early_stopping_rounds': trial.suggest_int('early_stopping_rounds', 10, 50)  
    }
    model = CatBoostRegressor(
        **params,
        loss_function='RMSE',
        eval_metric='RMSE',
        random_state=42,
        verbose=0
    )

    model.fit(X_train, y_train, eval_set=(X_val, y_val), early_stopping_rounds=params['early_stopping_rounds'])

    y_pred = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    
    return rmse

study = optuna.create_study(direction='minimize')

study.optimize(objective, n_trials=20, n_jobs=-1)  

print('Best hyperparameters:', study.best_params)
print('Best RMSE:', study.best_value)

best_params = study.best_params
catboost_model = CatBoostRegressor(
    **best_params,
    loss_function='RMSE',
    eval_metric='RMSE',
    random_state=42,
    verbose=200
)
catboost_model.fit(X_train, y_train, eval_set=(X_val, y_val))

y_pred = catboost_model.predict(X_val)

rmse = np.sqrt(mean_squared_error(y_val, y_pred))
print(f'Root Mean Squared Error (RMSE) on Validation Set: {rmse}')
```
